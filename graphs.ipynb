```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
psnr_fista = np.load('metricas/Fista_psnr.npy')
psnr_ista = np.load('metricas/Ista_psnr.npy')

plt.figure(figsize=(10, 6)) 

# 3. Graficar el PSNR
plt.plot(psnr_fista, color='b', label='FISTA') 
plt.plot(psnr_ista, color='r', label='ISTA')

plt.title('Algorithms comparison', fontsize=16)
plt.ylabel(r'PSNR (dB)', fontsize=14)
plt.xlabel(r'Iteration', fontsize=14)
plt.grid(True) 
plt.legend(fontsize=14)
plt.show()
```

```python
error_fista = np.load('metricas/Fista_error.npy')
error_ista = np.load('metricas/Ista_error.npy')

plt.figure(figsize=(10, 6))
# 3. Graficar el error
plt.plot(error_fista, color='b', label='FISTA')
plt.plot(error_ista, color='r', label='ISTA')

plt.title('Algorithms comparison', fontsize=16)
plt.yscale('log')
plt.ylabel(r'$\frac{1}{2} \|\mathbf{y} - \mathbf{H(x)}\|^2$', fontsize=14)
plt.xlabel(r'Iteration', fontsize=14)
plt.grid(True)
plt.legend(fontsize=14)
plt.show()
```
